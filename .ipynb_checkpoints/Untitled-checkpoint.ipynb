```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By

# Set up Chrome options for debugging mode
chrome_options = Options()
chrome_options.add_experimental_option("debuggerAddress", "localhost:9222")
driver = webdriver.Chrome(options=chrome_options)

# Run JavaScript to retrieve important elements on the page
elements = driver.execute_script("""
    let elements = [];
    document.querySelectorAll("a, button, input, textarea").forEach((el) => {
        elements.push({
            tag: el.tagName,
            id: el.id,
            className: el.className,
            text: el.innerText.trim(),
            placeholder: el.placeholder || null,
            href: el.href || null,
            name: el.name || null,
            type: el.type || null,
            value: el.value || null
        });
    });
    return elements;
""")

# Filter and display extracted elements based on importanc

# Function to filter and display only fields with values
# Function to filter elements and truncate long URLs
def filter_elements(elements):
    filtered_elements = []
    for element in elements:
        filtered_element = {}
        for key, value in element.items():
            # Skip empty or None values
            if value in [None, ''] or key == 'className'  :
                continue

             # Special case: If the element is an <A> tag with no text or href
            if key == 'tag' and value == 'A':
                # Check if both text and href are missing
                if not element.get('text') or not element.get('href'):
                    filtered_element = 0
                    break
            # If the value is a URL, truncate it to the first two words
            if key == 'href' and isinstance(value, str):
                # truncated_url = " ".join(value.split("/")[2:3]) + "..."  # Extract first two words of the URL
                # filtered_element[key] = truncated_url
                pass
            else:
                filtered_element[key] = value
          # Only append non-empty filtered elements
        if filtered_element:
            filtered_elements.append(filtered_element)
    return filtered_elements

# Filter the elements and print the result
filtered_elements = filter_elements(elements)
result = []
for element in filtered_elements:
    if element['tag'] == 'A' or element['tag'] == 'IMG' or element['type'] in ['button', 'submit'] or element['tag'] == 'TEXTAREA' or element['type'] == 'text':
        result.append(element)

print(result)
# for element in elements:
#     if element['tag'] == 'A' or element['type'] in ['button', 'submit'] or element['tag'] == 'TEXTAREA' or element['type'] == 'text':
#         print(element)
```
